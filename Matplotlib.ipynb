# Mini Project

```python
import matplotlib.pyplot as plt
```

```python
study_hrs = [1,6,7,9]
marks = [54,78,88,99]
```

# Line Plot

```python
plt.plot(study_hrs,marks, color = 'blue', marker = 'o')
plt.title("Students Performance")
plt.xlabel("Study Hrs")
plt.ylabel("Marks")
plt.show()
```

# Scatter Plot

```python
plt.scatter(study_hrs, marks, color='green', marker='x')
plt.title("Students Performance")
plt.xlabel("Study Hrs")
plt.ylabel("Marks")
plt.show()
```

# Bar plot

```python
plt.bar(study_hrs, marks, color='purple')
plt.title("Students Performance")
plt.xlabel("Study Hrs")
plt.ylabel("Marks")
plt.show()
```

# Horizontal Bar Plot

```python
plt.barh(study_hrs, marks, color='orange')
plt.title("Students Performance")
plt.xlabel("Marks")
plt.ylabel("Study Hrs")
plt.show()
```

# Histogram

```python
plt.hist(marks, bins=5, color='blue', edgecolor='black')
plt.title("Marks Distribution")
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.plot(study_hrs,marks,"o-r")
plt.title("Students Performance")
plt.xlabel("Study Hrs")
plt.ylabel("Marks")
plt.show()
```
