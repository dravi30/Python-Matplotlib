```python
import matplotlib.pyplot as plt
```

- Pixel (',')
- Circle ('o')

- Triangle Up ('^')
- Triangle Left ('<')
- Triangle Right ('>')

```python
sh = [1,2,3,4,5]
m = [56,76,87,99,100]
```

```python
#CIRCLE MARKER

plt.plot(sh,m,marker ='o')
plt.xlabel("Study Hours")
plt.ylabel("Marks Scored")
plt.show()
```

```python
#POINT MARKER
plt.plot(sh,m,marker ='.')
plt.xlabel("Study Hours")
plt.ylabel("Marks Scored")
plt.show()
```

```python
# pIXEL MARKER

plt.plot(sh,m,marker =',')
plt.xlabel("Study Hours")
plt.ylabel("Marks Scored")
plt.show()
```

```python
#Triangle Down Marker ('v')
plt.plot(sh,m,marker ='v',markersize=5)
plt.xlabel("Study Hours")
plt.ylabel("Marks Scored")
plt.show()
```

```python
#Triangle up Marker ('^')
plt.plot(sh,m,marker ='^',markersize=6)
plt.xlabel("Study Hours")
plt.ylabel("Marks Scored")
plt.show()
```

```python
plt.plot(sh, m, marker='s', markersize=6)
plt.xlabel("Study Hours")
plt.ylabel("Marks Scored")
plt.show()
# square marker
```

```python
plt.plot(sh, m, marker='*', markersize=6)
plt.xlabel("Study Hours")
plt.ylabel("Marks Scored")
plt.show()
# star marker
```

```python
plt.plot(sh, m, marker='D', markersize=6)
plt.xlabel("Study Hours")
plt.ylabel("Marks Scored")
plt.show()
# diamond 
```

```python
plt.plot(sh, m, marker='x', markersize=6)
plt.xlabel("Study Hours")
plt.ylabel("Marks Scored")
plt.show()
# x marker
```

```python
plt.plot(sh, m, marker='+', markersize=6)
plt.xlabel("Study Hours")
plt.ylabel("Marks Scored")
plt.show()

# + marker
```
