```python
# HISTOGRAM

import matplotlib.pyplot as plt
import numpy as np


x = np.random.normal(160,10,250)
print(x)
```

```python
plt.hist(x)
plt.show()
```

```python
#PIE CHART

x = np.array([20,34,75,90,85])
mylabel = ["A","B","c","D","E"]
separate = [0,0.2,0,0.2,0.2]
mycolors = ["red","blue","green","hotpink","lime"]

plt.pie(x,labels = mylabel, explode = separate,colors=mycolors)
plt.show()
```
